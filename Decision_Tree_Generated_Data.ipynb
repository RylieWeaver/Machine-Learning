```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import math
from sklearn.tree import plot_tree
```

```python
# data generation parameters
num_pts = 500
train_pts = 400
noise_lvl = 20
```

```python
# create a random array of features between -100 and 100
features = np.random.uniform(-100,100,num_pts)

# set the labels based on the value of each feature
labels = np.where(features < 0, 'A', 'B')

# create a dictionary to hold the data
data = {'Label': labels, 'Feature': features}

# create a pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# separate the data into features and labels
X_1 = df['Feature'].values.reshape(-1, 1)
y_1 = df['Label'].values

# split the data into training and testing sets
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.2)

# create a decision tree classifier
clf = DecisionTreeClassifier(max_depth=5)

# make X_train_1 have noise
noise1 = np.random.normal(0,noise_lvl,train_pts).reshape(-1,1)
X_train_1_noisy = X_train_1 + noise1

# fit the classifier to the training data
clf.fit(X_train_1_noisy, y_train_1)

# make predictions on the testing data
y_pred_1 = clf.predict(X_test_1)

# calculate the accuracy of the predictions
accuracy = accuracy_score(y_test_1, y_pred_1)

print(f'Accuracy: {accuracy:.2f}')

# plot the data points
fig, ax = plt.subplots(figsize=(10, 7))

markers = {'A': '^', 'B': 'o'}

for label in np.unique(y_1):
    ax.scatter(X_train_1_noisy[y_train_1==label], np.zeros_like(X_train_1[y_train_1==label]), marker=markers[label], label=label)

# plot the decision tree
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True)

ax.set_xlabel('Feature')
ax.set_ylabel('Label')
ax.legend()
plt.show()
```

## 2-Dimensional Dataset

```python
# create a generator to make features off of
generator = np.random.uniform(-10,10,num_pts)

# set the labels based on the generator value
labels = np.where(generator < 0, 'A', 'B')

# create 3 array of features based on the generator
feature_1 = np.log(np.absolute(generator))
feature_2 = generator**3

# create a dictionary to hold the data
data = {'Label': labels, 'Generator': generator, 'Feature 1': feature_1, 'Feature 2': feature_2}

# create a pandas DataFrame from the dictionary
df2 = pd.DataFrame(data)

# separate the data into features and labels
X_21 = df2['Feature 1'].values.reshape(-1, 1)
X_22 = df2['Feature 2'].values.reshape(-1, 1)
y_2 = df2['Label'].values

# combine features
X_2 = np.hstack([X_21, X_22])

# split the data into training and testing sets
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y_2, test_size=0.2)

# create a decision tree classifier
clf = DecisionTreeClassifier(max_depth=5)

# make X_train_2 have noise
noise2 = np.random.normal(0,noise_lvl,2*train_pts).reshape(-1,2)
X_train_2_noisy = X_train_2 + noise2

# fit the classifier to the training data
clf.fit(X_train_2_noisy, y_train_2)

# make predictions on the testing data
y_pred_2 = clf.predict(X_test_2)

# calculate the accuracy of the predictions
accuracy = accuracy_score(y_test_2, y_pred_2)

print(f'Accuracy: {accuracy:.2f}')

# plot the data points
fig, ax = plt.subplots(figsize=(10, 7))


markers = {'A': '^', 'B': 'o'}

for label in markers.keys():
    x = X_train_2_noisy[:, 0][y_train_2 == label]
    y = X_train_2_noisy[:, 1][y_train_2 == label]
    ax.scatter(x, y, label=label, marker=markers[label])

# plot the decision tree
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data Points with Labels')
plt.show()
```
