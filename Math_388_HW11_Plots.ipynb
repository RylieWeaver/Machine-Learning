```python
# imports

import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate F, H, and F2

F = np.arange(1,5.001,0.001)
# print(F)
H = (np.sqrt(1+8*F**2)-1)/2
# print(H)
F2 = F/(np.sqrt(H)**3)
# print(F2)
```

```python
# plot

plt.figure(figsize=(12,8))
plt.plot(F, H, label ='H')
plt.plot(F, F2, label ='F2') 
plt.legend()
plt.title("F vs H and F2 Plot")
plt.show()
```
